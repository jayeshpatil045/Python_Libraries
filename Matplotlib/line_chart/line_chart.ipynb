3.Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.

```python
def plot_line_from_file(filename):
    """
        Description:
            Reads x and y axis values from a text file and plots a line graph.

        Parameters:
            filename (str): The name of the text file containing the axis values.

        Returns:
            None
    """
    x = []
    y = []

    with open(filename, 'r') as file:
        for line in file:
            values = line.split()
            x.append(int(values[0]))
            y.append(int(values[1]))

    plt.plot(x, y)

    plt.xlabel('X Axis')
    plt.ylabel('Y Axis')

    plt.title('Line Plot from File Data')

    plt.show()

def main():
    filename = 'test.txt'
    
    plot_line_from_file(filename)

if __name__ == "__main__":
    main()
```

4.Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016.

```python
def read_financial_data(filename):
    """
        Description:
            Reads financial data from a CSV file.

        Parameters:
            filename (str): The path to the CSV file containing the financial data.

        Returns:
            dates (list of datetime): List of dates from the data.
            open_prices (list of float): List of 'Open' prices.
            high_prices (list of float): List of 'High' prices.
            low_prices (list of float): List of 'Low' prices.
            close_prices (list of float): List of 'Close' prices.
    """
    dates = []
    open_prices = []
    high_prices = []
    low_prices = []
    close_prices = []

    with open(filename, 'r') as fdata:
        reader = csv.reader(fdata)
        next(reader)  # Skip the header row
        for row in reader:
            date = datetime.datetime.strptime(row[0], '%d-%m-%y')
            dates.append(date)
            open_prices.append(float(row[1]))
            high_prices.append(float(row[2]))
            low_prices.append(float(row[3]))
            close_prices.append(float(row[4]))

    return dates, open_prices, high_prices, low_prices, close_prices

def plot_financial_data(dates, open_prices, high_prices, low_prices, close_prices):

    plt.figure(figsize=(10, 6))

    # Plotting each financial data
    plt.plot(dates, open_prices, label='Open', color='blue', marker='o')
    plt.plot(dates, high_prices, label='High', color='green', marker='o')
    plt.plot(dates, low_prices, label='Low', color='red', marker='o')
    plt.plot(dates, close_prices, label='Close', color='orange', marker='o')

    # Adding labels and title
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.title('Financial Data of Alphabet Inc. (October 3, 2016 - October 7, 2016)')

    plt.gcf().autofmt_xdate()

    plt.legend()

    # Display the plot
    plt.grid(True)
    plt.show()

def main():
    # Read financial data from CSV
    filename = 'fdata.csv'
    dates, open_prices, high_prices, low_prices, close_prices = read_financial_data(filename)

    # Plot the financial data
    plot_financial_data(dates, open_prices, high_prices, low_prices, close_prices)

if __name__ == "__main__":
    main()
```

9.Python program to display the current axis limits values and set new axis values.

```python
def display_and_set_axis_limits():
    """
        Description:
            Displays the current axis limits and sets new axis limits for a plot.

        Parameters:
            None

        Returns:
            None
    """
    x = [1, 2, 3, 4, 5]
    y = [10, 20, 25, 30, 40]

    plt.plot(x, y, marker='o')

    current_x_limits = plt.gca().get_xlim()
    current_y_limits = plt.gca().get_ylim()
    print(f"Current X axis limits: {current_x_limits}")
    print(f"Current Y axis limits: {current_y_limits}")

    plt.xlim(0, 6)  
    plt.ylim(5, 45) 

    plt.xlabel('X Axis')
    plt.ylabel('Y Axis')

    plt.title('Plot with Modified Axis Limits')
    plt.grid()
    plt.show()

def main():

    display_and_set_axis_limits()

if __name__ == "__main__":
    main()
```

10.Python program to plot quantities which have an x and y position.

12.Python program to create multiple types of charts

```python

def generate_data():
    """
        Description:
            Generates sample data for plotting.
        Parameters:
            None   

        Returns:
            x (array-like): X-axis values (equally spaced numbers).
            y1 (array-like): Y-axis values for the first line.
            y2 (array-like): Y-axis values for the second line.

    """
    x = np.linspace(0, 10, 100)  # 100 equally spaced numbers between 0 and 10
    y1 = np.sin(x)  # y1 is the sine of x
    y2 = np.cos(x)  # y2 is the cosine of x
    return x, y1, y2

def plot_line_chart(x, y1, y2):
   
    plt.plot(x, y1, label='Sine Wave', color='blue', linewidth=2)
    plt.plot(x, y2, label='Cosine Wave', color='green', linestyle='--', linewidth=2)
    
    # Adding labels and title
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Line Plot of Sine and Cosine Waves')
    
    # Adding a legend
    plt.legend()

    # Show grid
    plt.grid(True)
    
    # Display the plot
    plt.show()

def main():
    # Generate data for plotting
    x, y1, y2 = generate_data()

    # Plot the line chart
    plot_line_chart(x, y1, y2)

if __name__ == "__main__":
    main()
```

13.a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with linestyle -, width .5. and color blue.

```python
def read_closing_data(filename):
    """
        Description:
            Reads the closing value data from a CSV file.

        Parameters:
            filename (str): The path to the CSV file containing the closing value data.

        Returns:
            dates (list of datetime): List of dates from the data.
            close_prices (list of float): List of closing prices.
    """
    dates = []
    close_prices = []

    with open(filename, 'r') as fdata:
        reader = csv.reader(fdata)
        next(reader)  # Skip the header row
        for row in reader:
            date = datetime.datetime.strptime(row[0], '%d-%m-%y')
            dates.append(date)
            close_prices.append(float(row[1]))

    return dates, close_prices

def plot_closing_line_chart(dates, close_prices):

    plt.figure(figsize=(8, 5))

    # Plotting the closing prices
    plt.plot(dates, close_prices, label='Closing Price', color='green', marker='o', linestyle='-')

    # Customizing the grid
    plt.grid(True, which='both', linestyle='-', linewidth=0.5, color='blue')

    # Adding labels and title
    plt.xlabel('Date')
    plt.ylabel('Closing Price (USD)')
    plt.title('Closing Price of Alphabet Inc. (October 3, 2016 - October 7, 2016)')

    # Formatting the dates on x-axis
    plt.gcf().autofmt_xdate()

    # Display the plot
    plt.show()

def main():
    # Read the closing data from CSV
    filename = 'fdata.csv'
    dates, close_prices = read_closing_data(filename)

    # Plot the closing prices
    plot_closing_line_chart(dates, close_prices)

if __name__ == "__main__":
    main()
```

15.Write a Python program to create multiple plots.

```python
def create_multiple_plots():
    """
        Description:
            Creates multiple plots in a single figure: 
            - A sine wave 
            - A cosine wave
            - A quadratic function 
            - A cubic function
        Parameters:
            None
        Returns:
            None
    """

    # Generate x values from -10 to 10
    x = np.linspace(-10, 10, 100)
    
    # Compute y values for each plot
    y_sin = np.sin(x)
    y_cos = np.cos(x)
    y_quad = x ** 2
    y_cubic = x ** 3

    # Create a figure with 2 rows and 2 columns of subplots
    fig, axs = plt.subplots(2, 2, figsize=(10, 8))
    
    # Plot the sine wave
    axs[0, 0].plot(x, y_sin, 'r', label='Sine Wave')
    axs[0, 0].set_title('Sine Wave')
    axs[0, 0].legend()

    # Plot the cosine wave
    axs[0, 1].plot(x, y_cos, 'g', label='Cosine Wave')
    axs[0, 1].set_title('Cosine Wave')
    axs[0, 1].legend()

    # Plot the quadratic function
    axs[1, 0].plot(x, y_quad, 'b', label='Quadratic Function')
    axs[1, 0].set_title('Quadratic Function')
    axs[1, 0].legend()

    # Plot the cubic function
    axs[1, 1].plot(x, y_cubic, 'm', label='Cubic Function')
    axs[1, 1].set_title('Cubic Function')
    axs[1, 1].legend()

    # Add labels to all subplots
    for ax in axs.flat:
        ax.set(xlabel='X-axis', ylabel='Y-axis')

    # Adjust layout for better spacing
    plt.tight_layout()

    # Display the plots
    plt.show()

def main():
    # Create multiple plots
    create_multiple_plots()

if __name__ == "__main__":
    main()
```
