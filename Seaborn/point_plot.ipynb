```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

1.Write a program to draw a point plot for sex against survived for a dataset given in url
https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv

```python
def plot_point_sex_survived(url):
    """
        Description:
            Plots a point plot showing the survival probability by sex based on the Titanic dataset.

        Parameters:
            url (str): URL to the CSV file containing the Titanic dataset.

        Returns:
            None: Displays the point plot.
    """
    # Load the dataset from the URL
    titanic_data = pd.read_csv(url)

    # Create a point plot to visualize survival probability by sex
    sns.set(style="whitegrid")
    sns.pointplot(x="sex", y="survived", data=titanic_data)

    # Add labels and title
    plt.title('Survival Probability by Sex on the Titanic')
    plt.xlabel('Sex')
    plt.ylabel('Survival Probability')

    # Show the plot
    plt.show()

def main():
   
    # URL of the Titanic dataset
    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"

    # Call the function to generate the point plot
    plot_point_sex_survived(url)

# Entry point of the script
if __name__ == "__main__":
    main()
```
