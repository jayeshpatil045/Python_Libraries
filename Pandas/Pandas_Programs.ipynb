3.Write a Python program to add, subtract, multiple and divide two Pandas Series

```python
import pandas as pd

def arithmetic_operation(data1, data2):
    """
        Description:
            Perform basic arithmetic operations (addition, subtraction, multiplication, and division) between two pandas Series.

        Parameters:
            data1 (pd.Series): The first pandas Series containing numeric values.
            data2 (pd.Series): The second pandas Series containing numeric values.

        Returns:
            tuple: A tuple containing four pandas Series corresponding to the addition, subtraction,multiplication, and division of the two input Series.
    """
    addition = data1 + data2
    subtraction = data1 - data2
    multiplication = data1 * data2
    division = data1 / data2
    return addition, subtraction, multiplication, division

def main():
    
    data1 = pd.Series([2, 4, 6, 8, 10])
    data2 = pd.Series([1, 3, 5, 7, 9])
    print(f"The Given Series are:\n{data1}\nand\n{data2}")  
    
    added_series, subtracted_series, multiplied_series, divided_series = arithmetic_operation(data1, data2)
    
    print("Addition:\n", added_series)
    print("Subtraction:\n", subtracted_series)
    print("Multiplication:\n", multiplied_series)
    print("Division:\n", divided_series)

if __name__ == "__main__":
    main()
```

17.Python program to append a new row 'k' to data frame with given values for each column. Now delete the new row and return the original DataFrame.

```python
def append_and_remove_row(df, row_label, row_data):
    """
        Description:
            Appends a new row to the DataFrame and then deletes it, returning the original DataFrame.

        Parameters:
            df : The original DataFrame.
            row_label : The label for the new row.
            row_data : A dictionary containing the data for the new row.

        Returns:
            The original DataFrame after appending and removing the new row.
    """
    df.loc[row_label] = row_data
    
    print("DataFrame after appending the new row:")
    print(df, "\n")
    
    df = df.drop(row_label)
    
    return df

def main():

    exam_data = {
        'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael',
                 'Matthew', 'Laura', 'Kevin', 'Jonas'],
        'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
        'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
    }
    labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

    df = pd.DataFrame(exam_data, index=labels)

    new_row_label = 'k'
    new_row_data = {
        'name': 'Suresh',
        'score': 15.5,
        'attempts': 1,
        'qualify': 'yes'
    }

    original_df = append_and_remove_row(df, new_row_label, new_row_data)

    print("Original DataFrame after removing the new row:")
    print(original_df)

if __name__ == "__main__":
    main()
```

21.Python program to insert a new column in existing DataFrame.

```python
def insert_new_column(df, column_name, data, position=None):
    """
        Description:
            Inserts a new column into the DataFrame at the specified position.

        Parameters:
            df (pd.DataFrame): The DataFrame where the new column will be inserted.
            column_name (str): The name of the new column to be added.
            data (list): The data for the new column.
            position (int, optional): The position where the new column should be inserted.
                                      If None, the column will be added at the end.

        Returns:
            The DataFrame with the new column added.
    """
    if position is not None:
        df.insert(position, column_name, data)
    else:
        df[column_name] = data
    return df

def main():
    exam_data = {
        'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael',
                 'Matthew', 'Laura', 'Kevin', 'Jonas'],
        'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
        'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
    }
    labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

    df = pd.DataFrame(exam_data, index=labels)

    new_column_data = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']

    updated_df = insert_new_column(df, 'grade', new_column_data)

    print("DataFrame after inserting new column 'grade':")
    print(updated_df)

if __name__ == "__main__":
    main()
```

23.Python program to get list from DataFrame column headers.

```python
def get_column_headers(df):
    """
        Description:
            Returns a list of column headers from the DataFrame.

        Parameters:
            df : The DataFrame from which to extract column headers.

        Returns:
            A list of column headers.
    """
    return list(df.columns)

def main():
    exam_data = {
        'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael',
                 'Matthew', 'Laura', 'Kevin', 'Jonas'],
        'score': [12.5, 9, 16.5, 15, 9, 20, 14.5, 18, 8, 19],
        'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
    }

    df = pd.DataFrame(exam_data)

    headers = get_column_headers(df)
    print("Column Headers:", headers)

if __name__ == "__main__":
    main()
```
