```python
import plotly.express as px
import pandas as pd
import numpy as np
```

1.Write a program to draw a scatter plot for random 1000 x and y coordinates

```python
def plot_scatter_random_points():
    """
        Description:
            Plots a scatter plot of 1000 random x and y coordinates using plotly.

        Parameters:
            None

        Returns:
            None
    """
    # Generate 1000 random x and y coordinates
    np.random.seed(0)  
    x = np.random.randn(1000)
    y = np.random.randn(1000)

    fig = px.scatter(x=x, y=y, title="Scatter Plot of 1000 Random Points", labels={"x": "X Coordinates", "y": "Y Coordinates"})

    fig.show()

def main():

    plot_scatter_random_points()

if __name__ == "__main__":
    main()
```

3.Write a program to draw a scatter plot for random 500 x and y coordinates and style it.

```python
def plot_styled_scatter():
    """
        Description:
            Plots a styled scatter plot for 500 random x and y coordinates.

        Parameters:
            None

        Returns:
            None
    """
    np.random.seed(0)  
    x = np.random.rand(500)
    y = np.random.rand(500)

    plt.figure(figsize=(8, 6))

    plt.scatter(x, y, 
                c='purple',         # Set color of the points
                s=50,               # Set size of the points
                edgecolor='black',   # Add black borders around the points
                alpha=0.7,           # Set transparency of points
                marker='o')          # Use circle markers

    # Add title and labels with custom font sizes and styles
    plt.title('Styled Scatter Plot of Random 500 Points', fontsize=16, fontweight='bold')
    plt.xlabel('X Coordinates', fontsize=12)
    plt.ylabel('Y Coordinates', fontsize=12)

    # Customize grid and background
    plt.grid(True, linestyle='--', alpha=0.6)  # Show grid with dashed lines
    plt.gca().set_facecolor('whitesmoke')      # Set light background color

    # Show the plot
    plt.show()

def main():
    """
    Main function to call the plot_styled_scatter function.

    Returns:
    None
    """
    plot_styled_scatter()

# Entry point of the script
if __name__ == "__main__":
    main()
```

4.Write a program to draw a scatter plot for a given dataset and show datalabels on hover
https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv

```python
def plot_scatter_with_hover_labels(url):
    """
        Description:
            Plots a scatter plot for the given dataset and shows data labels on hover.

        Parameters:
            url (str): URL to the CSV file containing the dataset.

        Returns:
            None.
    """
    data = pd.read_csv(url)

    fig = px.scatter(data, 
                     x='Population', 
                     y='Rank', 
                     hover_name='State',    # Show state name on hover
                     size='Population',     # Size of points based on population
                     title='Population vs Rank of USA States in 2014')

    # Customize the labels for x and y axes
    fig.update_layout(
        xaxis_title="Population",
        yaxis_title="Rank",
    )

    # Show the plot
    fig.show()

def main():

    url = "https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv"

    plot_scatter_with_hover_labels(url)

if __name__ == "__main__":
    main()
```
