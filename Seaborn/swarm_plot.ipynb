```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

1.Write a program to draw a swarm plot of total bill against size for a given dataset
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

```python
def plot_swarm_total_bill_by_size(url):
    """
        Description:
            Plots a swarm plot of 'total bill' against 'size' from the tips dataset.

        Parameters:
            url (str): URL to the CSV file containing the tips dataset.

        Returns:
            None: Displays the swarm plot.
    """
    tips_data = pd.read_csv(url)

    sns.set(style="whitegrid")
    sns.swarmplot(x="size", y="total_bill", data=tips_data)

    # Add labels and title
    plt.title('Swarm Plot of Total Bill by Size')
    plt.xlabel('Size of Group')
    plt.ylabel('Total Bill ($)')

    # Show the plot
    plt.show()

def main():

    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"

    # Call the function to generate the swarm plot
    plot_swarm_total_bill_by_size(url)

if __name__ == "__main__":
    main()
```

2.Write a program to draw swarm plot of “total bill” against day for a dataset given in url
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

```python
def plot_swarm_total_bill_by_day(url):
    """
        Description:
            Plots a swarm plot of 'total bill' against 'day' from the tips dataset.

        Parameters:
            url (str): URL to the CSV file containing the tips dataset.

        Returns:
            None: Displays the swarm plot.
    """
    # Load the dataset from the URL
    tips_data = pd.read_csv(url)

    # Create a swarm plot to visualize total bill vs day
    sns.set(style="whitegrid")
    sns.swarmplot(x="day", y="total_bill", data=tips_data)

    # Add labels and title
    plt.title('Swarm Plot of Total Bill by Day')
    plt.xlabel('Day of the Week')
    plt.ylabel('Total Bill ($)')

    # Show the plot
    plt.show()

def main():

    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"

    # Call the function to generate the swarm plot
    plot_swarm_total_bill_by_day(url)

# Entry point of the script
if __name__ == "__main__":
    main()
```
