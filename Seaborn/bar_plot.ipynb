```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

1.Write a program to draw bar plot of sex against survived for a dataset given in the url
https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv

```python
def plot_survival_by_sex(url):
    """
        Description:
            Plots a bar chart showing the survival probability by sex based on the Titanic dataset.

        Parameters:
            url (str): URL to the CSV file containing the Titanic dataset.

        Returns:
            None: Displays the bar plot.
    """
    # Load the dataset from the URL
    titanic_data = pd.read_csv(url)

    # Create a bar plot to visualize survival probability by sex
    sns.set(style="whitegrid")
    sns.barplot(x="sex", y="survived", data=titanic_data)

    # Add labels and title
    plt.title('Survival Probability by Sex on the Titanic')
    plt.xlabel('Sex')
    plt.ylabel('Survival Probability')

    # Show the plot
    plt.show()

def main():

    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"

    # Call the function to generate the plot
    plot_survival_by_sex(url)

if __name__ == "__main__":
    main()
```
