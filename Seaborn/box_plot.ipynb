```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

1.Write a program to draw box plot of life expectancy by continent for a data set given in aurl
https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv

```python
def plot_box_life_expectancy_by_continent(url):
    """
        Description:
            Plots a box plot of life expectancy by continent from the Gapminder dataset.

        Parameters:
            url (str): URL to the CSV file containing the Gapminder dataset.

        Returns:
            None: Displays the box plot.
    """
    gapminder_data = pd.read_csv(url)

    # Create a box plot to visualize life expectancy by continent
    sns.set(style="whitegrid")
    sns.boxplot(x="continent", y="lifeExp", data=gapminder_data)

    # Add labels and title
    plt.title('Box Plot of Life Expectancy by Continent')
    plt.xlabel('Continent')
    plt.ylabel('Life Expectancy (years)')

    # Show the plot
    plt.show()

def main():
 
    url = "https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv"

    plot_box_life_expectancy_by_continent(url)

if __name__ == "__main__":
    main()
```

2.Write a program to draw a box plot of day by tips for a dataset given in a url
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

```python
def plot_box_day_by_tips(url):
    """
        Description:
            Plots a box plot of 'day' by 'tips' from the tips dataset.

        Parameters:
            url (str): URL to the CSV file containing the tips dataset.

        Returns:
            None: Displays the box plot.
    """
    tips_data = pd.read_csv(url)

    sns.set(style="whitegrid")
    sns.boxplot(x="day", y="tip", data=tips_data)

    # Add labels and title
    plt.title('Box Plot of Tips by Day')
    plt.xlabel('Day of the Week')
    plt.ylabel('Tip Amount ($)')

    # Show the plot
    plt.show()

def main():
    """
    Main function to call the plot_box_day_by_tips function and provide the dataset URL.
    
    Returns:
    None
    """
    # URL of the tips dataset
    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"

    plot_box_day_by_tips(url)

if __name__ == "__main__":
    main()
```
