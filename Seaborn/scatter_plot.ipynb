```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

1.Write a program to draw a scatter plot of “day” against “total bill” for a dataset given in a url
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

```python
def plot_scatter_day_total_bill(url):
    """
        Description:
            Plots a scatter plot of 'day' against 'total bill' from the tips dataset.

        Parameters:
            url (str): URL to the CSV file containing the tips dataset.

        Returns:
            None: Displays the scatter plot.
    """
    # Load the dataset from the URL
    tips_data = pd.read_csv(url)

    # Create a scatter plot to visualize day vs total bill
    sns.set(style="whitegrid")
    sns.scatterplot(x="day", y="total_bill", data=tips_data)

    # Add labels and title
    plt.title('Scatter Plot of Total Bill by Day')
    plt.xlabel('Day of the Week')
    plt.ylabel('Total Bill ($)')

    # Show the plot
    plt.show()

def main():

    # URL of the tips dataset
    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"

    plot_scatter_day_total_bill(url)

if __name__ == "__main__":
    main()
```
