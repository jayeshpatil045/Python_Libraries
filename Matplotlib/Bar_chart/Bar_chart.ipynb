1.Python programming to display a bar chart of the popularity of programming Languages.

```python
def popularity_of_language(languages,Popularity):
    """ 
        Description:
            Plots a bar chart of the popularity of programming languages.

        Parameters:
            languages (list): A list of programming languages.
            popularity (list): A list of popularity percentages corresponding to the languages.

        Returns:
            None
    """

    plt.bar(languages,Popularity)

    plt.xlabel('Programming languages')
    plt.ylabel('Popularity')

    plt.title('Popularity of Programming Languages')
    plt.show

def main():
    languages= ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
    popularity_of_language(languages,popularity)

if __name__ == "__main__":
    main()

```

9.Python programming to display a bar chart of the popularity of programming Languages.

```python
def plot_language_popularity_with_margin(languages, popularity):
    """
        Description:
            Plots a bar chart of the popularity of programming languages with an increased bottom margin.

        Parameters:
            languages (list): A list of programming languages.
            popularity (list): A list of popularity percentages corresponding to the languages.

        Returns:
            None
    """
    plt.bar(languages, popularity, color='skyblue', edgecolor='blue', linewidth=2)

    plt.gcf().subplots_adjust(bottom=0.2)

    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity (%)')

    plt.title('Popularity of Programming Languages')

    plt.show()

def main():

    languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
    
    plot_language_popularity_with_margin(languages, popularity)

if __name__ == "__main__":
    main()
```

10.Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.

```python
def plot_scores_by(means_men, means_women):
    """
        Description:
            Plots a bar chart of scores by group and gender, using multiple X values on the same chart for men and women.

        Parameters:
            means_men (list): A list of mean scores for men.
            means_women (list): A list of mean scores for women.

        Returns:
            None
    """
    n_groups = len(means_men)

    index = np.arange(n_groups)

    bar_width = 0.35
    plt.figure(figsize=(9,9))

    plt.bar(index, means_men, bar_width, label='Men', color='blue')
    plt.bar(index +bar_width, means_women, bar_width, label='Women', color='pink')

    plt.xlabel('Group')
    plt.ylabel('Scores')

    plt.title('Scores by Group and Gender')

    plt.legend()

    plt.xticks(index + bar_width / 2, ('G1', 'G2', 'G3', 'G4', 'G5'))

    plt.show()

def main():

    means_men = [22, 30, 35, 35, 26]
    means_women = [25, 32, 30, 35, 29]

    plot_scores_by(means_men, means_women)

if __name__ == "__main__":
    main()
```

11.Write a Python program to create bar plot from a DataFrame.

```python
def plot_bars_from_dataframe(df):
    """
        Description:
            Plots a bar chart from a given DataFrame with the index as the labels on the x-axis.
        Parameters:
            df (DataFrame): A pandas DataFrame containing the data to be plotted.
        Return:
            None        
    """    
    df.plot.bar()
    plt.xlabel('Index')
    plt.ylabel('values')
    plt.title('Bar Plot from DataFrame')
    plt.xticks(rotation=0)
    plt.show()

def main():

    data = {
        'a': [2, 4, 6, 8, 10],
        'b': [4, 2, 4, 2, 4],
        'c': [8, 3, 7, 6, 4],
        'd': [5, 4, 4, 4, 3],
        'e': [7, 2, 7, 8, 3],
        'f': [6, 6, 8, 6, 2]
    }
    
    df = pd.DataFrame(data)

    plot_bars_from_dataframe(df)

if __name__ == "__main__":
    main()
```

12.Python program to create bar plots with error bars on the same figure.

```python
def plot_with_error_bars(categories,mean_velocity,std_dev_velocity):
    """
        Description:
            Plots a bar chart with error bars.
        Parameters:
            categories : A list file containing the categories.
            mean_velocity : A list containing the mean values.
            std_dev_velocity : A list containing the std_dev values.
        Return:
            None        
    """     
    
    x = np.arange(len(categories))
    plt.bar(x, mean_velocity, yerr=std_dev_velocity, capsize=5, color='lightblue', edgecolor='black', alpha=0.7)
    plt.xlabel('Samples')
    plt.ylabel('Mean Velocity')
    plt.title('Bar Plot with Error Bars for Mean Velocity')


    plt.xticks(x, categories)
    plt.show()

def main():
    categories = ['Sample 1', 'Sample 2', 'Sample 3', 'Sample 4']
    mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
    std_dev_velocity = [0.3314, 0.2278, 0.2836, 0.2645]

    plot_with_error_bars(categories,mean_velocity,std_dev_velocity)
if __name__=="__main__":
    main()
```

13.Python program to create bar plots with errorbars on the same figure. Attach a text label above each bar displaying men means (integer value).

```python
def plot_bar_with_error_bars(categories, mean_velocity, std_dev_velocity):
    """
        Description:
            Plots a bar chart with error bars.
        Parameters:
            categories : A list file containing the categories.
            mean_velocity : A list containing the mean values.
            std_dev_velocity : A list containing the std_dev values.
        Return:
            None        
    """       

    mean_velocity_int = [int(round(val)) for val in mean_velocity]

    x = np.arange(len(categories))

    bars = plt.bar(x, mean_velocity, yerr=std_dev_velocity, capsize=5, color='lightblue', edgecolor='black', alpha=0.7)

    plt.xlabel('Samples')
    plt.ylabel('Mean Velocity')
    plt.title('Bar Plot with Error Bars and Labels')

    plt.xticks(x, categories)

    for bar, label in zip(bars, mean_velocity_int):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height + 0.01, str(label), ha='center', va='bottom')

    plt.show()

def main():
    categories = ['Sample 1', 'Sample 2', 'Sample 3', 'Sample 4']
    mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
    std_dev_velocity = [0.3314, 0.2278, 0.2836, 0.2645]

    plot_bar_with_error_bars(categories, mean_velocity, std_dev_velocity)

if __name__ == "__main__":
    main()
```
